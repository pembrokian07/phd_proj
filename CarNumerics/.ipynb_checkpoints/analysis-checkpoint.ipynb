```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
#A = pd.read_csv('A.csv', header=None)
#B = pd.read_csv('b.csv', header=None)
raw_psi = pd.read_csv('x.csv', header=None)
print(raw_psi.shape)
#idx = pd.read_csv('neumann_indices.csv', header=None).apply(lambda x: int(x), axis=1)
```

```python
I=100
J=1000
da = (2*np.pi)/J
dr = 1/I

def get_psi(x,i):
    return x[(i-1)*J:i*J]

def plot_psi(psi):
    x = np.arange(len(psi))*360/J
    ax = sns.lineplot(data=psi.set_axis(x))
    ax.set_xlabel('Î±')
    ax.set_ylabel('$\psi$')
    return ax

def process_psi(x:pd.DataFrame, idx:pd.DataFrame = None) -> pd.Series:
    psi = pd.DataFrame(np.zeros((I*J, 1)))
    
    t_i = x.index.values
    if idx is not None:
        t_i[-idx.size:] = t_i[-idx.size] + idx
    
    return psi.merge(x.set_index(t_i), left_index=True, right_index=True, how='outer').loc[:,'0_y'].fillna(0)
    
psi = process_psi(raw_psi, idx=None)
```

```python
res = np.zeros(I)
for k in range(1,I+1):
    c_psi = get_psi(psi, k)
    ans = pd.DataFrame({'x':np.arange(len(c_psi))*360/J, 
                    'psi':[np.power((k*dr),2)*(np.sin(x*da) + np.cos(x*da)) for x in range(1,J+1)]})
    res[k-1] = np.mean(np.abs(c_psi - [np.power(((k)*dr),2)*(np.sin(x*da) + np.cos(x*da)) for x in range(1,J+1)]))
    
errs = pd.Series(res)
```

## plot slns

```python
def check_sln(psi, idx):
    c_psi = get_psi(psi, idx)
    J = c_psi.shape[0]
    da = 2*np.pi/J
    
    psi_2 = c_psi.values
    s1_psi_2 = c_psi.shift(1).fillna(c_psi.values[-1]).values
    s2_psi_2 = c_psi.shift(-1).fillna(c_psi.values[0]).values
    
    psi_1 = get_psi(psi, idx-1).values
    psi_3 = get_psi(psi, idx+1).values
    
    d2psi_dr2 = (psi_3+psi_1-2*psi_2)/(dr**2)
    dpsi_dr = (psi_3 - psi_1)/(2*dr)
    d2psi_da2 = (s1_psi_2+s2_psi_2-2*psi_2)/(da**2)
    
    alpha = (np.arange(J)+1)*2*np.pi/J
    real_soln = 3*(np.sin(alpha) + np.cos(alpha))
    num_soln = d2psi_dr2 + dpsi_dr/(idx*dr) + d2psi_da2/np.power(idx*dr,2)

    return num_soln, real_soln
    
```

```python
def check_boundary(psi, idx):
    c_psi = get_psi(psi, idx)
    J = c_psi.shape[0]
    da = 2*np.pi/J
    
    psi_0 = c_psi.values
    m1_psi_0 = np.concatenate([psi_0[-1:],psi_0[0:-1]])
    m2_psi_0 = np.concatenate([psi_0[-2:],psi_0[0:-2]])
    p1_psi_0 = np.concatenate([psi_0[1:],psi_0[0:1]])
    p2_psi_0 = np.concatenate([psi_0[2:],psi_0[0:2]])
        
    psi_1 = get_psi(psi, idx-1).values
    psi_2 = get_psi(psi, idx-2).values
    psi_3 = get_psi(psi, idx-3).values
    psi_4 = get_psi(psi, idx-4).values
    
    dpsi_dr = (25/12*psi_0 - 4*psi_1 + 3*psi_2 - 4/3*psi_3 + 1/4*psi_4)/dr
    dpsi_da = (2/3*(p1_psi_0-m1_psi_0) - 1/12*(p2_psi_0-m2_psi_0))/da
    
    alpha = (np.arange(J)+1)*2*np.pi/J
    real_soln = idx*dr*(np.sin(alpha)*np.cos(alpha) + np.cos(alpha)**2 + 1)
    num_soln = np.cos(alpha)*dpsi_dr - np.sin(alpha)/(idx*dr)*dpsi_da
    
    return num_soln, real_soln
    
```

```python
num_soln, real_soln = check_boundary(psi, idx)
err = pd.DataFrame({'alpha':(np.arange(J)+1)*360/J, 'err': num_soln-real_soln})
```
